Looks like we currently have only one col that is technically numeric, the rating. however we'll most likely be able to change Reviews and Size and price to continuous variables unless it makes more sense to turn them into categorical variables and predict based on "bins"

Category, Type, Content Rating will definitely be our categorical variables.
Genres we'll eventually need to look at one-hotting everything as well. It's basically a bunch of T/Fs where each app can have one or more genres

Last Updted we'll change to a Date type. may or may not be useful. Probably apps that have NOT been update in a long time won't be a popular, whereas apps that have been updated any time in the fairly recent past will probably not vary too much based purely on the difference in date (i.e. 1 week ago and 2 weeks ago may not prove to be all that different on average, whereas over a year ago would be a red flag). don't know if it'll be necessary but it might be an option to group these last updateed dates into categories representing how (in)active the devs are.

Current version is probably a variable that acts as a proxy for other things we'd like to know, like how long hte app has been around and how often it's updated. So this actually could be useful, but we have to remember that we're not really measuring the impact of version #, which is technically arbitrary, but rather we're looking at a proxy (hopefully) for how active developers have been.

Android version we'll need to investigate a bit, but while having support for earlier versions could mean support for users and increased attractivity, this'll also be very highy correlated with how long the app's been around as well as things like current version and last updated. Newer apps may not have earlier support and that may not necessarrily be a damaging thing in reality, so we should keep this in mind as we do our analysis and model prep.



Variables that seem like they might be important off the bat if we have to choose now based on intuition:

-Reviews: more reviews tends to mean more popular, widely used apps. Also let's be honest you're more likely to download an app with 12000 testimonials than a similar looking one with 12 (or none).

-Rating: for obvious reasons. Let's keep in mind we want the ratings with higher sample size (more reviews) to be more important for our predictions

-Type (and Price): Type is free or paid so we expect free apps to be more popular in general, especially for some category like games, but paid apps may have higher quality. So obviously all else equal free>paid but that won't be the whole story. Same with price,, cheaper is generally better but there ought to be an opposite (but probably not equal) force that represent the fact that pricier apps tend** to have higher quality

-Size: I think this might be mixed. Smaller generally would be better all else equal but larger apps can potentially do more and have higher quality. Plus larger sizes just might be necessarry for certain categories. I think we'll see sizes grouped by app category in the data, so size alone might not be all that telling and may end up being a proxy for things like category and price



NOTE: if there seems to be great disparity in prediction from the genre and category and content rating variables, it might make sense to report findings by-category(, etc) as opposed to just using these variables as predictors. The reason being we want to inform app developers and marketers how to optimize their marketing, not what KIND OF APPS to produce. Predicting for each category (or group of categories could be far more useful for marketers in each category/grouping)

There are just 8 instances of nonstrings in this feature, and they're all NaN floats, so we can safely shunt them to their own invalid category in terms of creating major version feature

```python
import re

def get_major_ver(s):
    '''
    input s: string to parse
    returns: leading integers from the string if they exist, -999 if there's no
    match or match is > 20, or if input is not a string
    '''
    try:
        m = re.match(r'^[0-9]+', s)
        return int(m.group(0))
    except (AttributeError, TypeError):
        return np.nan
```

```python
df['Major_Ver'] = df['Current Ver'].apply(lambda x: get_major_ver(x))
df['Major_Ver'].value_counts()
```

Of course some of these major versions won't be helpful in any model, so let's change anything greater than '20' for now (though 20 might be higher than necessary tbh)

```python
df['Major_Ver'] = df['Major_Ver'].apply(lambda x: 21 if x > 20 else (1 if x == 0 else x))
```

```python
def convert_to_float(x):
    if 'Varies with device' in x:
        num = np.nan
    elif 'k' in x:
        num = float(x.replace('k', '')) * 1000 # convert k to a thousand
    elif 'M' in x:
        num = float(x.replace('M', '')) * 10**6 # convert M to a million
    elif 'B' in x:
        num = float(x.replace('B', '')) * 10**9 # convert B to a Billion
    else:
        num = int(x) # Less than 1000
    return float(num)
```

### Missing Values

```python
df_trimmed.isnull().sum()
```

We'll need to decide how to deal with missing values and NaNs in each column.

To start, let's drop rows with the target var Rating missing 

```python
df_trimmed = df_trimmed[df_trimmed['Rating'].notna()]
```

The remaining missings come from where version and size were marked as 'Varies with Device'. We don't want to drop all these rows since other features could still be useful. Instead I think it makes sense to impute values and maybe add a boolean column indicating that this app has variable version/size.

We've seen that Major Ver was mostly 1 and Min Android ver was overwhelmingly 4, so I'm going to impute by the mode for each of these for now.

```python
df_trimmed['Major_Ver'].fillna(value=1.0, inplace=True)
df_trimmed['Min_Android_Major_Ver'].fillna(value=4.0, inplace=True)
```

```python
def collapse_vals(i):
    if i >= 21:
        return '21+'
    elif i >= 9:
        return '9-20'
    else:
        return str(int(i))
```

```python
df_trimmed['Major_Ver'] = df_trimmed['Major_Ver'].apply(lambda x: collapse_vals(x))
df_trimmed['Major_Ver'].value_counts()
```

```python
df_trimmed['Major_Ver'] = pd.Categorical(df_trimmed['Major_Ver'], ordered=True)
df_trimmed['Min_Android_Major_Ver'] = pd.Categorical(df_trimmed['Min_Android_Major_Ver'], ordered=True)
```

As for Size_bytes, I think we'll impute by a mean/median grouped by other variable(s) like Android version, just to be a little more precise than imputing by the overall mean, which might be excessively large for many apps.

```python
df_trimmed.groupby('Major_Ver')['Size_bytes'].mean()
```

```python
df_trimmed.groupby('Min_Android_Major_Ver')['Size_bytes'].mean()
```

We see more variation with Android Min version

```python
df_trimmed.groupby('Install_Group')['Size_bytes'].mean()
```

But even more with Install_Group (lager apps generally mean larger size

```python
print('Difference: Overall InstallGroup proportions - NaN Size_bytes InstallGroup proportions: \n' \
      , df_trimmed['Install_Group'].value_counts() / len(df_trimmed) - \
      df_trimmed[df_trimmed['Size_bytes'].isna()]['Install_Group'].value_counts() / \
      len(df_trimmed[df_trimmed['Size_bytes'].isna()]))
```

What we can see from the above is that larger install groups make up a larger proportion of all data when Size is missing than they do when Size exists. (negative differences as we go to larger groups, though actually the last group is not the most negative so it's not a perfect decline).

But anyway I think it's worth trying to impute Size with mean grouped by install group, seeing as there is a clear difference between means across groups and the variables are correlated.

```python
df_trimmed['Size_bytes'] = df_trimmed.groupby('Install_Group')['Size_bytes'].transform(lambda x: x.fillna(x.mean()))
```
